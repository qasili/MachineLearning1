LINEAR REGRESSION

```python
X = df[['LATITUDE', 'LONGITUDE', 'LND_SQFOOT',
       'TOT_LVG_AREA', 'SPEC_FEAT_VAL', 'RAIL_DIST', 'OCEAN_DIST',
       'WATER_DIST', 'CNTR_DIST', 'SUBCNTR_DI', 'HWY_DIST', 'age', 'structure_quality']]
y = df[['SALE_PRC']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)

lnReg = LinearRegression() 
lnReg.fit(X_train, y_train)
```

```python
lnPredict = lnReg.predict(X_test) 
```

```python
# visualise predictions
plt.scatter(y_test,lnPredict) 
plt.xlabel('Actual Values') 
plt.ylabel('Predicted Values')
plt.show()
```

```python

# calculate the R Squared
r2 = r2_score(y_test, lnPredict)

# calculate the mean absolute error
mae = mean_absolute_error(y_test, lnPredict)

# calculate the RMSE
mse = mean_squared_error(y_test, lnPredict)  
rmse = math.sqrt(mse)

# calculate the MAPE
mape = mean_absolute_percentage_error(y_test, lnPredict)

# calculate the MdAE
mdAE = median_absolute_error(y_test, lnPredict)
```

```python
print(mae, rmse, mape, r2, mdAE)
```
